```python
import pandas as pd
import re

def clean_text(text):
    if not isinstance(text, str):
        return ""

    text = text.replace("’", "'").replace("‘", "'").replace("`", "'")

    text = re.sub(r'<.*?>', '', text)

    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    text = re.sub(r'\S+@\S+', '', text)

    text = re.sub(r'\d+', '', text)


    text = re.sub(r"[^a-zA-Z\s']", '', text)

    text = text.lower()

    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
df['text'] = df['text'].apply(clean_text)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X = df['text']
y = df['class'] 

print(f"Original X shape: {X.shape}")
print(f"Original y shape: {y.shape}")
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)

print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer(max_features=5000) # Re-add any parameters you used
XV_train = vectorization.fit_transform(X_train)
XV_test = vectorization.transform(X_test) # Use transform, not fit_transform for test set

print(f"Shape of XV_train after vectorization: {XV_train.shape}")
```

```python
print(XV_train)
print(y_train)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(XV_train,y_train)
```

Model Evaluation

```python
#Acuuracy of training data
XV_train_prediction=LR.predict(XV_train)
training_data_accuracy=accuracy_score(XV_train_prediction,y_train)
```

```python
print("Accuracy on training data :",training_data_accuracy)
```

```python
XV_test_prediction=LR.predict(XV_test)
test_data_accuracy=accuracy_score(XV_test_prediction,y_test)
```

```python
print("Accuracy on test data :",test_data_accuracy)
```
