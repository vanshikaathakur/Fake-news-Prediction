```python
import pandas as pd
import re

def clean_text(text):
    if not isinstance(text, str):
        return ""

    text = text.replace("’", "'").replace("‘", "'").replace("`", "'")

    text = re.sub(r'<.*?>', '', text)

    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    text = re.sub(r'\S+@\S+', '', text)

    text = re.sub(r'\d+', '', text)


    text = re.sub(r"[^a-zA-Z\s']", '', text)

    text = text.lower()

    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
import pandas as pd
import re
import unicodedata
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier # Assuming DT is Decision Tree
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier # Assuming GBC and RFC


def clean_text_with_apostrophes(text):
    """
    Cleans a given string by removing HTML tags, digits, most punctuation
    (but keeps apostrophes), extra whitespace, non-alphabetic characters,
    and converts to lowercase. Handles various Unicode characters.
    """
    if not isinstance(text, str):
        return ""

    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\S*@\S*\s?', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r"[^a-zA-Z\s']", '', text)
    text = text.lower()
    text = re.sub(r'\s+', ' ', text).strip()
    return text


def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News" 


sample_training_texts = [
    "this is a fake news article about politics",
    "scientists discover new planet it's true",
    "another fake story on social media",
    "real news report on economic growth"
]
dummy_vectorizer = TfidfVectorizer()
dummy_vectorizer.fit(sample_training_texts)
vectorization = dummy_vectorizer 
LR = LogisticRegression()
DT = DecisionTreeClassifier()
GBC = GradientBoostingClassifier()
RFC = RandomForestClassifier()

# Fit dummy models so they can be called (replace with your loaded/trained models)
dummy_X_train = vectorization.transform(sample_training_texts)
dummy_y_train = [0, 1, 0, 1] 

LR.fit(dummy_X_train, dummy_y_train)
DT.fit(dummy_X_train, dummy_y_train)
GBC.fit(dummy_X_train, dummy_y_train)
RFC.fit(dummy_X_train, dummy_y_train)


def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(clean_text_with_apostrophes)
    
 
    new_xv_test = vectorization.transform(new_def_test["text"])

    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GBC = GBC.predict(new_xv_test)
    pred_RFC = RFC.predict(new_xv_test)

    return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(
        output_label(pred_LR[0]),
        output_label(pred_DT[0]),
        output_label(pred_GBC[0]),
        output_label(pred_RFC[0])
    ))
    


news = str(input("Enter news text: ")) 
manual_testing(news)
```
